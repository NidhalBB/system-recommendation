```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Fake_dataset.csv',low_memory=False)
```

```python
df.groupby('Product Name')['Rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('Product Name')['Rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('Product Name')['Rating'].mean())
ratings['num of ratings'] = pd.DataFrame(df.groupby('Product Name')['Rating'].count())
ratings.head()
```

```python
moviemat = df.pivot_table(index='User ID',columns='Product Name',values='Rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['education']
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>2].sort_values('Correlation',ascending=False).head()
```

```python
liarliar_user_ratings = moviemat['pattern']
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>1].sort_values('Correlation',ascending=False).head()
```
